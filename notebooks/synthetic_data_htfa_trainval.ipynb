```python
import logging
import numpy as np
import htfa_torch.htfa as HTFA
import htfa_torch.niidb as niidb
import htfa_torch.utils as utils
```

```python
logging.basicConfig(format='%(asctime)s %(message)s', datefmt='%m/%d/%Y %H:%M:%S',
                    level=logging.INFO)
```

```python
synthetic_db = niidb.FMriActivationsDb('data/synthetic_data.db',
                                       mask='data/synthetic_data/nifti_files/wholebrain.nii.gz')
```

```python
htfa = HTFA.HierarchicalTopographicFactorAnalysis(list(synthetic_db.all()), mask='data/synthetic_data/nifti_files/wholebrain.nii.gz', num_factors=3)
```

```python
losses = htfa.train(num_steps=1000, learning_rate=1e-1, log_level=logging.INFO, num_particles=1,
                    batch_size=300, use_cuda=True, blocks_batch_size=9, checkpoint_steps=200,
                    blocks_filter=synthetic_db.inference_filter(training=True))
```

```python
utils.plot_losses(losses)
```

```python
losses = htfa.train(num_steps=1000, learning_rate=1e-1, log_level=logging.INFO, num_particles=1,
                    batch_size=300, use_cuda=True, blocks_batch_size=9, checkpoint_steps=200,
                    train_globals=False, blocks_filter=synthetic_db.inference_filter(training=False))
```

```python
utils.plot_losses(losses)
```

```python
for b in np.random.choice(range(htfa.num_blocks), 5):
    htfa.plot_factor_centers(b)
    htfa.plot_original_brain(b)
    htfa.plot_reconstruction(b)
    print(htfa.results(b))
```
