```python
import logging
import numpy as np
import htfa_torch.dtfa as DTFA
import htfa_torch.niidb as niidb
import htfa_torch.utils as utils
```

```python
%matplotlib inline
```

```python
logging.basicConfig(format='%(asctime)s %(message)s', datefmt='%m/%d/%Y %H:%M:%S',
                    level=logging.INFO)
```

```python
synthetic_db = niidb.FMriActivationsDb('data/synthetic_data_mini.db')
```

```python
dtfa = DTFA.DeepTFA([b for b in synthetic_db.all() if 'rest' not in b.task], mask='/home/zulqarnain/fmri_data/wholebrain.nii.gz', num_factors=3, embedding_dim=2)
```

```python
dtfa.tasks()
```

```python
dtfa.visualize_factor_embedding()
```

```python
losses = dtfa.train(num_steps=250, learning_rate={'q': 1e-1, 'p': 1e-2}, log_level=logging.INFO, num_particles=1,
                    batch_size=300, use_cuda=True, blocks_batch_size=20, patience=50, checkpoint_steps=200)
```

```python
utils.plot_losses(losses)
```

```python
for b in np.random.choice(range(dtfa.num_blocks), 5):
    dtfa.plot_factor_centers(b)
    dtfa.plot_original_brain(b)
    dtfa.plot_reconstruction(b)
    print(dtfa.results(b))
```

```python
def task_rest_labeler(task):
    if 'rest' in str(task):
        return 'Rest'
    else:
        return 'Task %s' % task
```
